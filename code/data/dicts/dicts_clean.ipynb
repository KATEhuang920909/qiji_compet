```python
import pandas as pd
data1=pd.read_excel("dics_df.xlsx")
data2=pd.read_excel(r"D:\code\qiji_compet\code\data\dicts\dics_df - 副本.xlsx")
```

```python
data1=data1[[0,1]].drop_duplicates(0,keep="first")
```

```python
data1.shape
```

```python
data_have=data1[0].values.tolist()
data_have
```

```python
add_data=[]
for a,b in data2.values:
    if a.strip() not in data_have:
        add_data.append((a.strip(),b))
```

```python
add_data=pd.DataFrame(add_data).drop_duplicates(0,keep="first")
add_data.shape
```

```python
data_dicts=pd.concat([data1,add_data]).drop_duplicates(0,keep="first")
```
